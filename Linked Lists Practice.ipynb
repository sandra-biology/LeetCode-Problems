```python
# reversing

head = None

node = ListNode(10)
node.next = head
head = node

node = ListNode(20)
node.next = head
head = node

node = ListNode(30)
node.next = head
head = node

node = ListNode(40)
node.next = head
head = node

node = head
while node:
    print(f"node.val = {node.val}")
    node = node.next

"""
Initialize three pointers prev as NULL, curr as head and next as NULL.
Iterate trough the linked list. In loop, do following.
// Before changing next of current,
// store next node
next = curr->next
// Now change next of current
// This is where actual reversing happens
curr->next = prev

// Move prev and curr one step forward
prev = curr
curr = next
"""

o_head = head
prev = None
curr = o_head
next = None
while curr:
    o_head = curr
    next = curr.next  # next keeps the tail of the linked list. Now we can change curr.next
    curr.next = prev

    prev = curr
    curr = next

print("\nafter reversing")
node = o_head
while node:
    print(f"node.val = {node.val}")
    node = node.next
```

```python
class LinkedList:
    class ListNode:
        def __init__(self, x):
            # ListNode constructor
            self.val = x
            self.next = None

    def __init__(self):
        # LinkedList constructor
        self.head = None
        self.tail = None
        
    def add_to_head(self, x):
        node = self.ListNode(x)
        node.next = self.head
        self.head = node
        if not self.tail:
            self.tail = node

    def add_to_tail(self, x):
        node = self.ListNode(x)
        self.tail.next = node
        self.tail = node

    def print(self):
        node = self.head
        while node:
            print(f"node.val = {node.val}")
            node = node.next


llist = LinkedList()

llist.add_to_head(40)
llist.add_to_head(30)
llist.add_to_head(10)
llist.add_to_tail(1000)
llist.add_to_tail(2000)

llist.print()
```
