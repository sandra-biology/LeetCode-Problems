###  List gets copied to get the output to be in the right order

```python
# passed all test cases

# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def mergeTwoLists(self, l1: ListNode, l2: ListNode) -> ListNode:
        head3 = None
        node1 = l1
        node2 = l2

        while node1 and node2:
            if node1.val < node2.val:
                # add node1 to head3
                node = ListNode(node1.val)
                node.next = head3
                head3 = node
                # increment node
                node1 = node1.next
            else:
                # add node2 to head3
                node = ListNode(node2.val)
                node.next = head3
                head3 = node
                # increment node
                node2 = node2.next

        while node1:
            # add node1 to output list
            node = ListNode(node1.val)
            node.next = head3
            head3 = node
            node1 = node1.next

        while node2:
            # add node1 to output list
            node = ListNode(node2.val)
            node.next = head3
            head3 = node
            node2 = node2.next

        head4 = None
        node = head3
        while node:
            # create a new node with content of node.val
            node_copy = ListNode(node.val)
            # add node to the list
            node_copy.next = head4
            head4 = node_copy
            node = node.next

        return head4


solution = Solution()
res = solution.mergeTwoLists(head1, head2)

node = res
while node:
    print(f"output node.val = {node.val}")
    node = node.next
```
