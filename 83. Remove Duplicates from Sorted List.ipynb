```python
from collections import defaultdict

class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def deleteDuplicates(self, head: ListNode) -> ListNode:
        # create dictionary to keep track of number of times each node appears in list
        node_dict = defaultdict(int)
        # go through linked list
        node = head
        while node:
        # add first node to dictionary
            if node.val not in node_dict:
                node_dict[node.val] += 1
                print(f"node_dict = {node_dict}")
                node = node.next
        # if subsequent node is not in dictionary, add it
        ## otherwise, remove it from list (node.next becomes node.next.next)
            else:
                node.next = node.next.next
        return head

solution = Solution()

head = None

node = ListNode(2)
node.next = head
head = node

node = ListNode(1)
node.next = head
head = node

node = ListNode(1)
node.next = head
head = node

res = solution.deleteDuplicates(head)

node = res
while node:
    print(f"output node.val = {node.val}")
    node = node.next
```
