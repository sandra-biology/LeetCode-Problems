```python
class Queue:
    def __init__(self):
        self.queue = []
    
    def push(self, x):
        self.queue.append(x)
    
    def pop(self):
        return self.queue.pop()
    
    def top(self):
        return self.queue[-1]
    
    def empty(self):
        return len(self.queue) == 0

queue = Queue()  # creates object
queue.push(10)
queue.push(20)
queue.push(30)

print(f"queue.queue = {queue.queue}")
queue.pop()
print(f"after pop(): queue.queue = {queue.queue}")

print(f"queue.top() = {queue.top()}")
print(f"queue.empty() = {queue.empty()}")
```

```python
class MyStack:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.stack = Queue()

    def push(self, x: int) -> None:
        """
        Push element x onto stack.
        """
        self.stack.push(x)

    def pop(self) -> int:
        """
        Removes the element on top of the stack and returns that element.
        """
        return self.stack.pop()

    def top(self) -> int:
        """
        Get the top element.
        """
        return self.stack.top()

    def empty(self) -> bool:
        """
        Returns whether the stack is empty.
        """
        return self.stack.empty()
    
    def get_container(self):
        return self.stack.queue


stack = MyStack()
stack.push(40)
stack.push(50)
stack.push(60)
print(f"stack.mystack = {stack.stack.queue}")
# print(f"stack.mystack = {stack.get_container()}")
stack.pop()
print(f"stack after pop: stack.mystack = {stack.stack.queue}")

print(f"stack.top() = {stack.top()}")
print(f"stack.empty() = {stack.empty()}")
```
